**Removing Skewness**

🎯 Why Remove Skewness?

Highly skewed data can:

1. Affect the performance of machine learning models

2. Mislead statistical analysis (especially mean-based stats)

3. Break assumptions of algorithms like Linear Regression or Logistic Regression



```python
from sklearn.preprocessing import PowerTransformer
import numpy as np
numeric_df = df.select_dtypes(include=['number'])
skewed_cols = numeric_df.columns[numeric_df.skew().abs() > 1]
for col in skewed_cols:
    if col in df.columns:
        df[col] = np.log1p(df[col]) if (df[col] > 0).all() else PowerTransformer(method='yeo-johnson').fit_transform(df[[col]])

print("Skewness After Transformation:\n", df[skewed_cols].skew())
```

**Feature selection**

  Feature selection is the process of choosing the most important columns (features) from your dataset that contribute the most to your prediction or analysis task.

  Instead of using all the columns, you pick only the most useful ones — which helps your model become faster, more accurate, and less confusing.

```python
from sklearn.compose import ColumnTransformer
from sklearn.feature_selection import SelectKBest,f_classif
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
#Feature selection using SelectKBest
x=df.drop(columns='Temperature_F',axis=1)
y=df['Temperature_F']
categorical_features=x.select_dtypes(include=['object']).columns
numerical_features=x.select_dtypes(exclude=['object']).columns
categorical_transformer=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
numerical_transformer='passthrough'
preprocessor=ColumnTransformer(
    transformers=[
        ('num',numerical_transformer,numerical_features),
        ('cat',categorical_transformer,categorical_features)
    ]
)
x_encoded=preprocessor.fit_transform(x)
x_encoded_df=pd.DataFrame(x_encoded,columns=preprocessor.get_feature_names_out())
selecter=SelectKBest(score_func=f_classif,k=4)
x_new=selecter.fit_transform(x_encoded_df,y)
selected_features=x_encoded_df.columns[selecter.get_support()]
feature_score=pd.DataFrame({'features':x_encoded_df.columns,'score':selecter.scores_}).sort_values(by='score',ascending=False)
feature_score
print("Selected features")
print(selected_features)
print("\nFeature scores")
print(feature_score)
```

Why Do We Use Feature Selection?

   Feature selection is important because it helps improve the efficiency, accuracy, and clarity of machine learning models. By selecting only the most relevant features, we can build models that are faster, less complex, and more resistant to overfitting. It removes unnecessary data that might act as noise, saving both time and computational resources while making the results easier to interpret.

**Split Training and Testing Data**

What is Split Training and Testing Data?
Splitting data means dividing your dataset into two parts:

1. Training set – used to train the machine learning model.

2. Testing set – used to evaluate how well the model performs on unseen data.

This is usually done in a ratio like 80% training / 20% testing or 70/30.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
#Split data into training and testing
x_train,x_test,y_train,y_test=train_test_split(x_encoded_df,y,test_size=0.2,random_state=42)
```

Why Do We Split Data into Training and Testing Sets?

Splitting data into training and testing sets is a key step in building effective machine learning models. It helps the model learn patterns from the training set and then evaluates performance on unseen data using the test set. This process ensures the model is not just memorizing data (overfitting), but is actually able to generalize and make accurate predictions in real-world scenarios.

**Scaling**

What is Scaling?

Scaling is the process of transforming the features (columns) in your dataset so that they all have similar ranges. It doesn’t change the data values, just adjusts the scale (like resizing).

```python
Ss=StandardScaler()
x_test_scaled=Ss.fit_transform(x_test)
x_train_scaled=Ss.fit_transform(x_train)
```

 Popular Scaling Techniques

There are different scaling methods used based on the nature of the data:

1.  StandardScaler: Best when data is normally distributed. It scales features to have a mean of 0 and a standard deviation of 1.

2. MinMaxScaler: Used when you want to scale features to a fixed range (usually 0 to 1). It’s sensitive to outliers.

3. RobustScaler: Ideal for data with outliers, as it uses the median and IQR instead of mean and standard deviation.

 Why is Scaling Important?

Scaling is essential in machine learning to ensure that all features contribute equally to the model. It prevents features with larger values from dominating the learning process. By bringing all features to a similar range, scaling improves model performance, ensures faster training (especially for gradient-based models), and makes distance calculations more accurate for clustering and other algorithms.

**Model performance**

Model performance refers to how well a machine learning model makes predictions on new, unseen data. It is evaluated using various metrics that tell us how accurate, reliable, and effective the model is.



```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
y_train_discrete = (y_train > 0).astype(int)
y_test_discrete = (y_test > 0).astype(int)
for name,clf in classifiers.items():
    clf.fit(x_train_scaled,y_train_discrete) # Use the discrete target for training
    y_pred=clf.predict(x_test_scaled)
    print(f"\n{name}Results:")
    print("Accuracy:",accuracy_score(y_test_discrete,y_pred)) # Use the discrete target for evaluation
    print("\nConfusion Matrix:")
    print(confusion_matrix(y_test_discrete,y_pred))
    print("\nClassification Report:")
```
